```python
for j in [1,18,19,21,22,25]:
    for i in range(0, 205):
        try:
            eval(df.iloc[i,j])
        except:
            
        # print("The wrong value is :",df2.iloc[i,0], "at index number",i)
            df.replace(df.iloc[i,j],np.nan,inplace=True)
```

# Univariate Analysis

#  Price Histogram + KDE

```python
sns.histplot(df['price'],color='royalblue', kde=True)
plt.title('Price Distribution')
plt.show()
```

##### Price Distribution: The market is right-skewed, meaning most cars are budget-friendly ($5,000–$12,000), while a few luxury outliers pull the average up.


# Make (Brand) Frequency

```python
sns.countplot(data=df, y='make',palette='viridis')
plt.title('Frequency of Vehicle Makes')
plt.show()
```

##### Brand Frequency: Toyota heavily dominates the dataset, followed by other Japanese brands like Nissan and Mazda, indicating a focus on mass-market reliability.

# Body Style Frequency

```python
sns.countplot(data=df, x='body-style',palette='magma')
plt.title('Frequency of Body Styles')
plt.show()
```

##### Body Style: Sedans and Hatchbacks make up the vast majority of the inventory, reflecting a consumer-driven data set focused on practical, everyday transportation.

# Bivariate Analysis

## Heatmap

```python
df_num = df.select_dtypes([float, int])
corr = df_num.corr()
corr
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='mako')
plt.title('Feature Correlations')
plt.show()
```

##### Identifies a "synergistic cluster" where physical size, engine capacity, and horsepower move in tight synchronization, creating a multi-dimensional profile of high-value vehicles.

# Engine Size Regression

```python
sns.regplot(data=df, x='engine-size', y='price')
```

##### Engine Size vs. Price: Exhibits a dominant positive correlation (0.87), establishing engine displacement as the primary mechanical determinant of vehicle valuation.

## Body Style Boxplot

```python
sns.boxplot(data=df, x='body-style', y='price')
plt.show()
```

##### The Body Style analysis reveals that while hardtop and convertible models command the highest median market premiums, the sedan and hatchback segments exhibit significant price volatility characterized by high-end luxury outliers.

# Multivariate Analysis

## Engine Size vs Price (Segmented by Fuel Type)

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")


sns.scatterplot(data=df, x='engine-size', y='price', 
                hue='fuel-type', style='fuel-type', s=100, palette='hot')


sns.regplot(data=df[df['fuel-type']=='gas'], x='engine-size', y='price', scatter=False, color='purple', label='Gas Trend')
sns.regplot(data=df[df['fuel-type']=='diesel'], x='engine-size', y='price', scatter=False, color='royalblue', label='Diesel Trend')

plt.title('How Fuel Type affects the Engine-Price Relationship', fontsize=14)
plt.xlabel('Engine Size', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.legend()
plt.show()
```

##### The regression analysis demonstrates a linear elasticity between engine displacement and market value, with diesel powertrains showing a marginal price premium and higher cost-sensitivity to engine scaling compared to gasoline models.

## Price Distribution by Body Style & Drive Wheels

```python
plt.figure(figsize=(12, 7))


sns.boxplot(data=df, x='body-style', y='price', hue='drive-wheels', palette='viridis')

plt.title('Price Variation: Body Style + Drive System', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##### The analysis identifies rear-wheel drive (rwd) as the definitive premium configuration, consistently elevating price floors and medians across all body styles compared to the more economy-focused fwd and 4wd counterparts.

## Mean Price Heatmap (The Pivot Table Logic)

```python
pivot_data = df.pivot_table(values='price', index='drive-wheels', columns='body-style', aggfunc='mean')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data, annot=True, fmt=".0f", cmap='YlGnBu', linewidths=.5)

plt.title('Average Price Heatmap: Who is the most expensive?', fontsize=14)
plt.show()
```

##### The visualization confirms that rear-wheel drive (rwd) acts as the primary price catalyst across all body styles, consistently commanding a significant premium over fwd and 4wd configurations.
