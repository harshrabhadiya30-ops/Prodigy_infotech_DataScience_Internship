```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python

df_pop = pd.read_csv('TimelineData.csv', skiprows=4)
df_meta = pd.read_csv('Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_127039.csv')


merged_df = pd.merge(df_pop, df_meta, on='Country Code')
df_pop.head()

```

# Categorical Visualization: Bar Chart


```python
plt.figure(figsize=(10, 6))
sns.countplot(data=merged_df, y='Region', palette='flare', order=merged_df['Region'].value_counts().index)
plt.title('Distribution of Countries by Region')
plt.xlabel('Count of Countries')
plt.ylabel('Region')
plt.show()


```

##### Insights : The Pareto distribution of the metadata reveals that approximately 25% of the world's reporting entities are located in Europe & Central Asia, making it the most granular region in this dataset. Conversely, North America represents the smallest category, indicating that while it is a major economic driver, it provides the least amount of individual data points for categorical comparison.


# Continuous Visualization: Histogram


```python
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['2023'].dropna(), bins=30, kde=True, color='slateblue')
plt.title('Global Population Distribution (2023)')
plt.xlabel('Population (Log Scale)')
plt.ylabel('Frequency')
plt.xscale('log') 
plt.show()
```

##### Insights : The chart shows that world population is extremely lopsided. While a few giant countries like India and China have huge populations, most countries in the world actually have fewer than 20 million people. This tells us that the 'average' population size is misleading because it's being pushed up by just a few massive nations, while the majority of the world's countries are actually quite small in comparison
